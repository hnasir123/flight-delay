Get average passenger numbers for each route for each month


```python
wanted_cols = ['unique_carrier', 'airline_id', 'unique_carrier_name', 'origin', 'dest', 'seats', 'passengers','year', 'month']
avg_passengers_df = passengers_df[wanted_cols]
avg_passengers_df['percent_full'] = round(avg_passengers_df['passengers'] / avg_passengers_df['seats'] * 100, 2)
avg_passengers_df.head(10) 
```

```python
avg_passengers_df[(avg_passengers_df['airline_id'] == 20409) & (avg_passengers_df['month']==4) & (avg_passengers_df['origin']=='MCO') & (avg_passengers_df['dest']=='NAS')]
```

```python
avg_passengers_df['route'] = avg_passengers_df['origin'] + ' to ' + avg_passengers_df['dest']
avg_passengers_df.head()
```

```python
avg_passengers_df['route'].nunique()
```

```python
avg_passengers_df[avg_passengers_df['origin'] == avg_passengers_df['dest']]['origin'].unique()
```

```python
route_month_avg_dict = avg_passengers_df.groupby(['route', 'month'])['passengers'].mean().to_dict()
```

```python
avg_passengers_df['map_key'] = list(zip(avg_passengers_df['route'], avg_passengers_df['month']))
```

```python
avg_passengers_df['avg_passengers'] = avg_passengers_df['map_key'].map(route_month_avg_dict)
```
